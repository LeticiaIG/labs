```python
# FILTRAR PALABRA POR LAS QUE TERMINAN EN ING
    #palabras > todas las palabras de la lista
    
def busqueda_ing (words):
    lista = []
    for w in words:
        if 'ing' in w:
            lista.append(w)
    else:
        pass
    return lista

#Buscamos actividad, buscamos -ing
palabras = busqueda_ing (palabra)
print(palabras)
print(type(palabras))
# o, eliminar las palabras comunes
    # c.subtract(d)
```
