## 2. renaming columns and skipping rows

### Date

```python
# extracting months from "date" column

# months:
df['Date'] = df["Date"].replace('Jan','01',regex =True)
df['Date'] = df["Date"].replace('Feb','02',regex =True)
df['Date'] = df["Date"].replace('Mar','03',regex =True)
df['Date'] = df["Date"].replace('Apr','04',regex =True)
df['Date'] = df["Date"].replace('May','05',regex =True)
df['Date'] = df["Date"].replace('May','05',regex =True)
df['Date'] = df["Date"].replace('Jun','06',regex =True)
df['Date'] = df["Date"].replace('Jul','07',regex =True)
df['Date'] = df["Date"].replace('Aug','08',regex =True)
df['Date'] = df["Date"].replace('Sep','09',regex =True)
df['Date'] = df["Date"].replace('Oct','10',regex =True)
df['Date'] = df["Date"].replace('Nov','11',regex =True)
df['Date'] = df["Date"].replace('Dec','12',regex =True)
# removing non numeric characters and any whitespace character:
df['Date'] = df["Date"].replace('[a-zA-Z]','',regex =True)
df['Date'] = df["Date"].replace('\s','',regex =True)
# r years:
df['Date'] = df["Date"].replace('\d\d\d\d','',regex =True)
# r days:
df['Date'] = df["Date"].replace('^\d+-','',regex =True)
# r points , 0 , -
df['Date'] = df["Date"].replace('\.','',regex =True)
df['Date'] = df["Date"].replace('\-','',regex =True)
df['Date'] = df["Date"].replace('^\d{3,}','',regex =True)
df['Date'] = df["Date"].replace('0','',regex =True)
df['Date'] = df["Date"].replace('','1',regex =True)


df['Date'].unique()
df['Date'].value_counts()
```

```python
#r values != 1-12
def getMonth(*month):
    if month[0]=="":
        return 0
    elif int(month[0])>12:
        return 0
    return int(month[0])

df['Date']= df['Date'].apply(getMonth)


#removing rows if value = 0
df = df.drop(df[df['Date'] == 0].index, axis=0)
df['Date'].unique()

#renaming column
df = df.rename(columns={'Date': 'MONTH'})
df.head()
```

```python
def ChangeYears(years):
    if int(years) >= 0 and int(years) <= 14: return "Children"
    elif int(years) > 14 and int(years) <= 24: return "Youth"
    elif int(years) > 24 and int(years) <= 35: return "Young Adult"
    elif int(years) > 35 and int(years) <= 64: return "Adult"
    elif int(years) > 64 and int(years) <= 99: return "Senior"
    else: return "Null"
    
df['Age'] = df['Age'].apply(ChangeYears)
df['Age'].value_counts()

df = df.rename(columns={"Age":"AGE"})
df.head()
```

```python
# ----------------------------    UNFINISHED     -------------------------


df['Injury'].value_counts()

#cómo para que sea siempre la primera que encuentre?
def InjuryClassify(injury):
    if 'no injury' in str(injury).lower(): return 'NO INJURY, SLIGHT SCARE!'
    if 'foot' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'feet' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'hip' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'hips' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'leg' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'toe' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'toes' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'knee' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'knees' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'ankle' in str(injury).lower(): return 'LOWER EXTREMITIES'
    if 'ankles' in str(injury).lower(): return 'LOWER EXTREMITIES'
    
    if 'fatal' in str(injury).lower(): return 'FATAL'
    else: return "OTHERS"

    
df['Injury'] = df['Injury'].apply(InjuryClassify)
df = df.rename(columns={'Injury':'INJURY'})
df['INJURY'].value_counts()   
```
